2.1 Program #1

>   ```c#
>       string s1 = "I \u0003 Visual C# ";
>       System.Console.WriteLine($"s1 = {s1}");
>       
>       // 1. เมธอด IndexOf() ใช้ค้นหาลำดับที่ของอักขระที่ปรากฏใน string และส่งค่าตำแหน่งที่พบมาให้ (Zero index)
>       var indexOfC = s1.IndexOf("C");
>       System.Console.WriteLine($"Index of C in s1 = {indexOfC}");
>       
>       // 2. เมธอด Substring(  ) คัดลอกบางส่วนของ string ต้นฉบับออกมา โดยการระบุพารามิเตอร์ >    ตำแหน่ งเริ่มต้นและความยาวที่ต้องการ 
>       var substring = s1.Substring(indexOfC, 2);
>       System.Console.WriteLine($"s1.Substring({indexOfC}, 2) = {substring}");
>       
>       // 3. เมธอด Replace(  ) ใช้แทนที่บางส่วนของ string  ด้วย string ที่กำหนด 
>       System.Console.WriteLine($"Replace C# with Basic and write to s2");
>       System.Console.WriteLine($"Before replace s1.hashcode = {s1.GetHashCode():X}"); 
>       var s2 = s1.Replace("C#", "Basic");
>       System.Console.WriteLine($"s1 = {s1}");
>       System.Console.WriteLine($"s2 = {s2}");
>       System.Console.WriteLine($"After replace s1.hashcode = {s1.GetHashCode():X}"); 
>   ```

2.2 อธิบายโปรแกรม

ให้ศึกษาโปรแกรมด้านบนและอธิบายการทำงาน

string s1 = "I \u0003 Visual C# ";
สร้างตัวแปรสตริงชื่อ s1 และกำหนดค่าเริ่มต้นเป็น "I \u0003 Visual C# " โดยมีอักขระพิเศษ \u0003 แทรกอยู่ (อาจเป็นอักขระควบคุม)

System.Console.WriteLine($"s1 = {s1}");
แสดงค่าของ s1 ออกทางคอนโซล

var indexOfC = s1.IndexOf("C");
ใช้เมธอด IndexOf() เพื่อค้นหาตำแหน่งแรกที่พบอักขระ 'C' ในสตริง s1 และเก็บค่าตำแหน่งไว้ในตัวแปร indexOfC
System.Console.WriteLine($"Index of C in s1 = {indexOfC}");
แสดงตำแหน่งที่พบ

var substring = s1.Substring(indexOfC, 2);
ใช้เมธอด Substring() เพื่อดึงสตริงย่อยจาก s1 เริ่มต้นที่ตำแหน่ง indexOfC (ที่พบอักขระ 'C') และมีความยาว 2 ตัวอักษร
System.Console.WriteLine($"s1.Substring({indexOfC}, 2) = {substring}");
แสดงสตริงย่อยที่ได้

System.Console.WriteLine($"Replace C# with Basic and write to s2");
แจ้งว่าจะทำการแทนที่สตริง

var s2 = s1.Replace("C#", "Basic");
ใช้เมธอด Replace() เพื่อสร้างสตริงใหม่ s2 โดยแทนที่ "C#" ใน s1 ด้วย "Basic" แต่สตริง s1 เดิมยังคงไม่เปลี่ยนแปลง
System.Console.WriteLine($"s1 = {s1}");
แสดงค่าของ s1 อีกครั้งเพื่อยืนยันว่ายังคงเหมือนเดิม
System.Console.WriteLine($"s2 = {s2}");
แสดงค่าของ s2 ที่ได้จากการแทนที่
System.Console.WriteLine($"After replace s1.hashcode = {s1.GetHashCode():X}");
แสดงค่า hash code ของ s1 อีกครั้งหลังจากการแทนที่ (ค่า hash code จะเปลี่ยนไปเนื่องจากเป็นสตริงใหม่)

3.2 อธิบายโปรแกรม

ให้ศึกษาโปรแกรมด้านบนและอธิบายการทำงาน

string s1 = "I \u0003 Visual C# ";

สร้างสตริง s1 ที่มีค่าเริ่มต้นคือ "I Visual C# " โดยมีอักขระพิเศษ \u0003 แทรกอยู่

System.Console.WriteLine($"s1 = {s1}");

แสดงค่าของสตริง s1 ออกทางคอนโซล

var indexOfC = s1.IndexOf("C");

ใช้เมธอด IndexOf เพื่อค้นหาตำแหน่งแรกที่พบอักขระ 'C' ในสตริง s1 และเก็บค่าตำแหน่งนั้นไว้ในตัวแปร indexOfC

var substring = s1.Substring(indexOfC, 2);

ใช้เมธอด Substring เพื่อตัดตอนสตริง s1 เริ่มต้นจากตำแหน่งที่พบอักขระ 'C' (ซึ่งได้จาก indexOfC) และมีความยาว 2 ตัวอักขระ ผลลัพธ์จะเก็บไว้ในตัวแปร substring

s1 = s1.Replace("C#", "Basic");

ใช้เมธอด Replace เพื่อแทนที่ทุก occurrences ของ "C#" ในสตริง s1 ด้วย "Basic" และเก็บผลลัพธ์กลับไปที่ตัวแปร s1 เดิม

System.Console.WriteLine($"Before replace s1.hashcode = {s1.GetHashCode():X}");

แสดงค่าแฮชของสตริง s1 ก่อนการแทนที่ โดยใช้ GetHashCode() เพื่อแปลงสตริงเป็นค่าจำนวนเต็ม และ :X เพื่อแสดงค่าในรูปแบบเลขฐานสิบหก
