## String interpolation

String interpolation ในภาษา C# เป็นวิธีการที่ สะดวกและอ่านง่าย ในการฝังค่าของตัวแปรลงใน string โดยตรง

__วิธีการใช้งาน__

1. ใส่เครื่องหมาย $ ไว้หน้า string: นี่เป็นการบอกคอมไพเลอร์ว่าเราต้องการใช้ string interpolation
2. ใส่ตัวแปรหรือ expression ในเครื่องหมายปีกกา {}: คอมไพเลอร์จะแทนที่ expression นี้ด้วยค่าที่ผ่านการคำนวณ

ทดลองรันโปรแกรมต่อไปนี้

````
string name = "John";
int age = 30;

string message = $"My name is {name} and I am {age} years old.";

Console.WriteLine(message); 
````

#### ข้อดีของ String Interpolation
- __อ่านง่าย__ โค้ดดูสะอาดตาและเข้าใจง่ายกว่าการใช้ string concatenation (+)
- __สะดวก__  ไม่ต้องแปลงชนิดข้อมูลด้วยตนเอง คอมไพเลอร์จะจัดการให้เอง
- __ยืดหยุ่น__ สามารถใส่ expression ใดๆ ก็ได้ใน {} เช่น การคำนวณ การเรียกใช้เมธอด ฯลฯ

เราสามารถใส่ string  formatter รวมทั้งการคำนวณเอาไว้ใน string interpolation ได้

````
double price = 19.99;
int quantity = 3;

string order = $"You ordered {quantity} items at {price:C} each. Total: {price * quantity:C}";

Console.WriteLine(order);
````

3. อธิบายโปรแกรม

ให้ศึกษาโปรแกรมด้านบนและอธิบายการทำงาน

var student = (firstname: "fname", Lastname:"lname", id: 65031234);

สร้างตัวแปรแบบ tuple: สร้างตัวแปร student ที่เป็นชนิด tuple ซึ่งเป็นโครงสร้างข้อมูลที่ใช้เก็บข้อมูลหลายชนิดในตัวแปรเดียว โดยแต่ละข้อมูลจะมีชื่อ (เช่น 
firstname, lastname, id)

กำหนดค่าเริ่มต้น: กำหนดค่าเริ่มต้นให้กับแต่ละส่วนของ tuple โดยใช้ชื่อที่กำหนดไว้ เช่น firstname เป็น "fname", lastname เป็น "lname", และ id เป็น 65031234

แก้ไขข้อมูล: ให้แก้ไขค่าของ firstname, lastname และ id ให้เป็นข้อมูลของนักศึกษาจริง

System.Console.WriteLine(
@$"First name: {student.firstname} 
Last name: {student.Lastname} 
ID:{student.id}");

แสดงผลลัพธ์: แสดงข้อมูลของนักศึกษาที่แก้ไขแล้วออกมาทาง console โดยใช้ string interpolation (ใส่ค่าตัวแปรเข้าไปในสตริงโดยตรง)

int x = 3;
int y = 4;
System.Console.WriteLine($$"""Distance from origin to the point ({{x}},{{y}}) is {{Math.Sqrt(x*x+y*y)}}""");

กำหนดค่าพิกัด: กำหนดค่าของพิกัด x และ y

คำนวณระยะทาง: ใช้สูตรคณิตศาสตร์ในการหาระยะทางจากจุดกำเนิด (0, 0) ไปยังจุด (x, y) โดยใช้ฟังก์ชัน Math.Sqrt เพื่อหาค่ารากที่สอง

แสดงผลลัพธ์: แสดงระยะทางที่คำนวณได้ออกมาทาง console โดยใช้ string interpolation แบบ raw string literal ใส่ $$ ก่อนและหลังสตริง เพื่อให้สามารถใส่ { และ } ได้โดยตรง)

System.Console.WriteLine("\n\nMethod to show { and } in string");
System.Console.WriteLine($$"""Distance from origin to the point ({x},{y}) is {Math.Sqrt(x*x+y*y)}""");

แสดงวิธีการ: แสดงข้อความเพื่อบอกว่าส่วนนี้จะแสดงวิธีการใส่ { และ } ในสตริง

ใช้ raw string literal: ใช้ raw string literal เพื่อให้สามารถใส่ { และ } ในสตริงได้โดยตรงโดยไม่ต้อง escape
