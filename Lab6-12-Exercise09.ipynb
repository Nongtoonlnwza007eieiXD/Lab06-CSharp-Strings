# Lab 3 Exercise 9



## Using StringBuilder
จากการทดลองที่ผ่านมา เราจะเห็นว่า string เป็น object ชนิด immutable คือไม่สามารถแก้ไขได้ การแก้ไขใดๆ ใน string จะส่งผลให้ระบบสร้าง string ใหม่ขึ้นมาแทน string เก่า

ในการใช้งานที่มีปริมาณการกระทำต่อ string ซ้ำ ๆ เป็นจำนวนมาก อาจจะก่อให้เกิดปัญหาด้านประสิทธิภาพการทำงานของระบบคอมพิวเตอร์ได้

คลาส StringBuilder จะมีการสร้าง buffer ของ string ไว้ภายใน ทำให้ไม่ต้องสร้าง string ใหม่เมื่อมีการเปลี่ยนแปลงค่าในข้อความดังเช่นการใช้ string โดยทั่วไป

3. อธิบายโปรแกรม

ให้ศึกษาโปรแกรมด้านบนและอธิบายการทำงาน


System.Text.StringBuilder sb = new System.Text.StringBuilder("Rat is my favorite animal.");: ส่วนนี้สร้างอ็อบเจ็กต์ StringBuilder ชื่อ sb และกำหนดค่าเริ่มต้นเป็น "Rat is my favorite animal." StringBuilder จะเก็บสตริงไว้ในหน่วยความจำในลักษณะที่สามารถแก้ไขได้โดยไม่ต้องสร้างอ็อบเจ็กต์ใหม่

System.Console.WriteLine(@$"Before change : sb = {sb.ToString()}, sb.GetHashCode = {sb.GetHashCode()}");: ส่วนนี้พิมพ์ค่าปัจจุบันของ sb และค่าแฮช (hash code) ของอ็อบเจ็กต์ StringBuilder ออกทางคอนโซล

sb.ToString(): แปลง StringBuilder เป็น string เพื่อให้สามารถแสดงผลได้
sb.GetHashCode(): คืนค่าแฮชของอ็อบเจ็กต์ ค่าแฮชใช้สำหรับเปรียบเทียบความเท่ากันของอ็อบเจ็กต์อย่างรวดเร็ว โดยทั่วไปแล้ว ถ้าอ็อบเจ็กต์สองตัวมีค่าแฮชที่แตกต่างกัน แสดงว่าอ็อบเจ็กต์เหล่านั้นไม่เท่ากัน แต่ถ้าอ็อบเจ็กต์สองตัวมีค่าแฮชเท่ากัน ไม่ได้หมายความว่าอ็อบเจ็กต์เหล่านั้นจะเท่ากันเสมอไป (อาจเกิด collision ได้)
System.Console.WriteLine(@"Replace first char in sb from R to C");: ส่วนนี้พิมพ์ข้อความอธิบายว่าจะทำการเปลี่ยนแปลงอะไร

sb[0] = 'C';: ส่วนนี้คือส่วนสำคัญที่แสดงความแตกต่างระหว่าง string และ StringBuilder ส่วนนี้ทำการแก้ไขอักขระตัวแรกของ sb จาก 'R' เป็น 'C' เนื่องจาก StringBuilder เป็นแบบเปลี่ยนแปลงได้ การแก้ไขนี้เกิดขึ้นโดยตรงกับอ็อบเจ็กต์เดิม ไม่มีการสร้างอ็อบเจ็กต์ใหม่

System.Console.WriteLine(@$"After change : sb = {sb.ToString()}, sb.GetHashCode = {sb.GetHashCode()}");: ส่วนนี้พิมพ์ค่าของ sb หลังจากการเปลี่ยนแปลง และค่าแฮชใหม่ออกทางคอนโซล

6. อธิบายโปรแกรม

ให้ศึกษาโปรแกรมด้านบนและอธิบายการทำงาน

var sb = new System.Text.StringBuilder();: บรรทัดนี้ประกาศตัวแปร sb ซึ่งเป็นอ็อบเจ็กต์ของคลาส StringBuilder คลาสนี้ใช้สำหรับสร้างและแก้ไขสตริงแบบไดนามิก ซึ่งมีประสิทธิภาพมากกว่าการใช้ string โดยตรงเมื่อมีการแก้ไขสตริงหลายครั้ง

for (int i = 0; i < 10; i++): ลูป for นี้จะวนซ้ำ 10 ครั้ง โดย i เริ่มต้นที่ 0 และเพิ่มขึ้นทีละ 1 จนถึง 9

System.Console.Write ($"Loop {i+1, 2} : ");: บรรทัดนี้แสดงผลข้อความ "Loop" ตามด้วยหมายเลขลูป (เริ่มจาก 1 ถึง 10) โดยใช้การจัดรูปแบบ {i+1, 2} เพื่อให้ตัวเลขมีอย่างน้อย 2 หลัก เช่น "Loop 1", "Loop 2", ..., "Loop 10"

sb.Append(i.ToString());: บรรทัดนี้เป็นส่วนสำคัญ sb.Append() ใช้สำหรับเพิ่มสตริงเข้าไปใน StringBuilder ในที่นี้ i.ToString() แปลงค่าของ i (ซึ่งเป็นตัวเลข) ให้เป็นสตริง แล้วเพิ่มเข้าไปใน sb ในแต่ละรอบของลูป ตัวเลขจะถูกต่อกันไปเรื่อยๆ

System.Console.WriteLine($"sb = {sb.ToString(),-11}, sb.GetHashCode() = {sb.GetHashCode()}");: บรรทัดนี้แสดงผลค่าของ sb ในปัจจุบัน และ HashCode ของ sb

sb.ToString(): แปลง StringBuilder sb เป็น string เพื่อแสดงผล
{sb.ToString(),-11}: ใช้การจัดรูปแบบเพื่อให้สตริงที่แสดงผลมีความกว้างอย่างน้อย 11 ตัวอักษร และชิดซ้าย (เครื่องหมายลบ -) เพื่อให้การแสดงผลเป็นระเบียบ
sb.GetHashCode(): คืนค่า Hash Code ของอ็อบเจ็กต์ StringBuilder ซึ่งเป็นค่า Integer ที่ใช้ในการเปรียบเทียบอ็อบเจ็กต์ (โดยทั่วไปแล้ว Hash Code จะเปลี่ยนไปเมื่อเนื้อหาของอ็อบเจ็กต์เปลี่ยนไป)
Console.WriteLine($"Finally : sb = {sb}");: หลังจากลูป for ทำงานเสร็จสิ้น บรรทัดนี้จะแสดงผลค่าสุดท้ายของ sb ซึ่งจะเป็นสตริงที่ประกอบด้วยตัวเลข 0 ถึง 9 ที่ต่อกัน
